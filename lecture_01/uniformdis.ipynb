- Write a piece of code that will sample 10,000 points uniformly distributed on the surface of a sphere. Your code should return the points' angular coordinates $\theta, \phi$ (you can assume a unit sphere wlog). 

```python
import numpy as np

def generate_uniform_points_on_sphere(num_points):
    # Generate random points in spherical coordinates
    phi = 2 * np.pi * np.random.rand(num_points)  # Azimuthal angle (longitude)
    costheta = 2 * np.random.rand(num_points) - 1  # Cosine of polar angle (colatitude)
    theta = np.arccos(costheta)  # Polar angle (latitude)

    # Return the spherical coordinates
    return theta, phi

# Generate 10,000 uniformly distributed points on the unit sphere
num_points = 10000
theta, phi = generate_uniform_points_on_sphere(num_points)

```

```python
import matplotlib.pyplot as plt

plt.scatter(phi, theta,s=1)
```

- Write a piece of code that will project the points into Carthesian coordinates ($x, y, z$), again assuming unit radius.

```python
def transfromer(theta,phi):
    x = np.sin(theta)*np.cos(phi)
    y = np.sin(theta)*np.sin(phi)
    z = np.cos(theta)
    return x, y, z
```

```python
theta, phi = generate_uniform_points_on_sphere(10000)
x,y,z = transfromer(theta, phi)
```
