## $\texttt{Python}$ basics

What is $\texttt{python}$?
* $\texttt{Python}$ is an interpreted language, which makes it easy to learn and use.
* $\texttt{Python}$ heavily relies on modules (libraries) which already contain a vast number of ready-to-use functions; once these modules are installed they are ready to use!

### Syntax
In contrast to other coding languages, indentations play an important role in $\texttt{python}$ and are not just to make your code more reader friendly. Indentation denotes the beginning of a new code block (for example for a if-statement, function definition, ...). Brackets are not necessary. For the indentation, you can either use the tab or white spaces. It doesn’t matter where you put white spaces (apart from the very left-most white space).

### Declaring Variables
In $\texttt{python}$ we do not need to declare the type of the variable. We simply define a variable by assigning a value:

```python
# integer
a = 1
b = 10
print(a, b)
print(type(a), type(b))

# float 
c = 1.0 
d = 1e10 
print(c, d)
print(type(c), type(d))

# string
e = "a"
f = "hello world" 
print(e, f)
print(type(e), type(f))

# bool
g = True
h = False
print(g, h)
print(type(g), type(h))
```

$\texttt{Python}$ automatically assigns the correct type. It is even possible to reassign a value of a different type:

```python
# integer
a = 1

# now a will be a float 
a = 1.1
```

### Data Types
**List:**

A simple collection of elements is a list. In $\texttt{python}$ is is generated by using squared barackets:

```python
# Creating a simple list
list_1 = [1,2,3]

# The elements in the list do not need to have the same types 
list_2 = [1,"a",[1,2,3]]

# Creating an empty list
list_empty = []
```

We can extract individual elements from the list by using the index numbers:

```python
# Creating a simple list
list_1 = [1,2,3,4]

# The first element: 
print("The first element:", list_1[0])

# The last element
print("The last element:", list_1[-1]) 

# A slice of elements
print("A slice of elements:", list_1[1:3]) 
```

We can also reassign elements in the list or append new ones.

```python
# Creating a simple list
list_1 = [1,2,3,4]

# Reassign
list_1[0] = 100

# Append a new element 
list_1.append("new element") 

# Print the new list 
print(list_1)
```

**Dictionary:**
    
These are similar to lists, but they are unordered. They are written using curly brackets and have keys and values :

```python
# Defining a simple dictionary
dict_1 = { "key_1": 1, 
           "key_2": 2}
```

The elements have the keys key 1 and key 2 which have the values 1, and 2 assigned to them. We can access the elements from a specific key in the following way:

```python
print(dict_1["key_1"])
```

We can extract the full information from a dictionary using the build in `.keys()`, `.values()` and `.items()` functions:

```python
# The dictionary keys
print("Keys: ", dict_1.keys())

# The attribute values
print("Values: ", dict_1.values())

# A list of the items in the dictionary 
print("Items: ", dict_1.items())
```

### Basic Operations
When working with integer or floats, the simple operations are `+,-,*,/,% and **`:

```python
a = 1
b = 2
c = 3.0

# Simple addition 
print("Addition: ", a+b)

# An operation with an integer and a float will output a float 
print("Integer and float: ", a+c)

# Simple subtraction 
print("Subtraction: ", a-b)

# Simple division will always return a float output 
print("Division: ", b/a)

# To achieve integer division, use // 
print("Integer division: ", b//a)

# Taking the square of a number 
print("Square: ", b**2)

# The remainder of a number: 
print("Remainder: ", b%a)

# Adding strings 
c = "Hello "
d = "world!" 
print("Adding strings: ", c+d)
```

### If–Statements
With an if–statement, we can condition what the script should execute and what not. The following conditions are allowed:
- Equal: `a == b`.
- Not Equals: `a!=b`.
- Less than; Greater than: `a<b` or `a>b`.
- Less than or equal; Greater than or equal: `a<=b` or `a>=b`.

Notice that we have to indent the lines after the if-statement to indicate which lines belong to the statement:

```python
a = 1 
b = 2
if a == b:
    print("a equals b")
```

We can use the elif and else statement to further specify our conditioning statement:

```python
a = 1 
b = 2
if a == b:
    print("a equals b") 
elif a > b:
    print("a greater b") 
else:
    print("b greater a")
```

### For–Loops
A for loop iterates over a sequence and performs a task for the number of iterations. By indentation, $\texttt{python}$ knows which lines of code belong to the for loop. Notice that we do not need to set an index variable beforehand, which is the case in other programming languages. A simple for loop is generated the following way:

```python
items = ["Astro","is","cool!"] 

for item in items:
    print(item)
```

Often times, the `range()` function is used to iterate a task a certain number of times:

```python
# This loop will run from 0 to 2. This gives us a total number of repetitions of 3
for i in range(3): 
    print(i)
```

```python
# Range with a different step size: 
for i in range(0, 6, 2):
    print(i)
```

### Functions
We can define a simple function using the `def` command:

```python
def f(a):
    print("You entered ", a)
    
f("hello")

f(1)
```

Often we want to assign the output. We can return an output of a function using `return`:

```python
def f(x): 
    return 2*x

a = f(2) 
print(a)
```

A function can also have multiple input parameters. It is also customary to comment in the header of the function, what the input parameters need to be:

```python
def power(x, n): 
    """
    Computes the nth power of x.
    Args:
    x (:obj:`integer`, `float` or `array`): base
    n (:obj:`integer` or `float`): exponent
    
    Returns:
        (:obj:`integer`, `float` or `array`): the exponential
    """
    
    return x**n

result = power(2, 3) 
print(result)
```

A function can also have optional arguments:

```python
def sum_three(a, b, c=0):
    """
    Computes the sum of two or three numbers.
    Args:
    a (:obj:`integer` or `float`): first number to sum
    b (:obj:`integer` or `float`): second number to sum
    c (:obj:`integer` or `float`): third number to sum, default c=0
    
    Returns:
        (:obj:`integer`, `float` or `array`): the sum of a,b,c
    """
    d = a + b + c
    
    return d
```

```python
a = 3
b = 15
sum_three(3, 15)
```

```python
a = 3
b = 15
c = 5
sum_three(3, 15, 5)
```

Multiplication of two 2 dimensional arrays is still elementwise, it is not matrix multiplication.

One thing to keep in mind with $\texttt{python}$ objects: object assignment creates a reference to the same object:

```python
arr1 = np.arange(1, 5)
arr2 = arr1
print('First array')
print(arr1)
print('Second array')
print(arr2)
arr1[1] = 1000
print('First array')
print(arr1)
print('Second array')
print(arr2)
```

In order to force python to create a new object we need to copy the array:

```python
arr1 = np.arange(1, 5)
arr2 = np.copy(arr1)
print('First array')
print(arr1)
print('Second array')
print(arr2)
arr1[1] = 1000
print('First array')
print(arr1)
print('Second array')
print(arr2)
```

**Array masking:**

Because $\texttt{python}$ is an interpreted language, especially the heavy use of for loops can make your code really slow. Especially when dealing with large arrays, it might be of an advantage to use masks as opposed to for loops.
Suppose we have an 1 D array a with a number of float values. If we want to set all negative values to zero, we can do that either using a for loop:

```python
a = np.random.uniform(-10, 10, 1000000)
print(a)
```

```python
%%time
for i in range(len(a)): 
    if a[i] < 0:
        a[i] = 0
```

This will set all negative values to zero. However, if we have a very large array, this can take quite some time. A better way is using a mask the following way:

```python
%%time
a[a<0] = 0
```

This directly sets all negative values to zero and is computationally much faster than using a for loop.

### Plotting in $\texttt{python}$
When we are creating plots and graphs with $\texttt{python}$, we usually use the module `matplotlib`. Is has many built-in functions to customize the graphs and figures. The following script will create a simple, basic graph:

```python
import matplotlib.pyplot as plt 
import numpy as np

# Create the data
x = np.linspace(0, 5, 100)

# Plot the data
plt.plot(x, x, label="linear line", linewidth=2, color='#AE7182')
# Add a legend
plt.legend()
# Add axes labels
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
# Show the plot
# plt.show()
```

We can plot several lines or objects into the same figure:

```python
# Prepare the data
x = np.linspace(0, 5, 100)

# Plot the data
plt.plot(x, x, label="linear", linewidth=2, color='#AE7182') 
plt.plot(x, x**2, label="squared", linewidth=2, color='#725e9c') 
plt.plot(x, x**3, label ="cubed", linewidth=2, color='#e3a19c')
# Add a legend
plt.legend()
# Add axes labels
plt.xlabel(r"$x$")
plt.ylabel(r"$y$")
# Show the plot
# plt.show()
```

## $\texttt{Git}$ basics

### What is $\texttt{git}$?

$\texttt{Git}$ is a version control system, i.e. it tracks changes in files and thus makes it very easy to (i) go back to previous versions (absolutely crucial :-)), (ii) collaborate with other people.

$\texttt{Git}$ runs locally, and all your files and their history are stored on your computer. However, you can also use online hosts (we will use [GitHub](https://github.com/)) to store a copy of the files and their revision history. Having a centrally located place where you can upload your changes and download changes from others, facilitates collaboration with others. $\texttt{Git}$ has a number of tools to facilitate merging the changes from several people: For example two people can work on different parts of the same file and later merge those changes without losing each other’s work.
